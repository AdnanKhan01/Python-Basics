```python
import numpy as np
a  = np.random.randint(1,100, size=(1000,5))
df = pd.DataFrame(a,
                 columns=["s1","s2","s3","s4","s5"])
df
```

```python
df['total']= df['s1']+df['s2']+df.s3+df['s4']+df.s5
df
```

```python
def my_addition(s1, s2,s3,s4,s5):
  return s1 + s2 + s3 + s4+ s5

df['obtained'] = df[["s1","s2","s3","s4","s5"]].apply(lambda x:my_addition(*x), axis=1)
df
```

```python
df[["s1","s2","s3","s4","s5"]].apply(sum) #top to bottom
```

```python
def my_addition(s1, s2,s3,s4,s5):
  obt =  s1 + s2 + s3 + s4+ s5
  total = 500
  per = (obt / total) * 100
  grade = "I"

  if per >= 80:
    grade = "A+"
  elif per >= 70:
    grade = "A"
  elif per >= 60:
    grade = "B"
  elif per >= 50:
    grade = "C"
  else:
    grade = "Fail"


  return total, obt, per, grade
#Passing result_type=’expand’ will expand list-like results to columns of a Dataframe:

df[["total","obtained","per","grade"]] =df[["s1","s2","s3","s4","s5"]].apply(lambda x: my_addition(*x),axis=1, result_type='expand')
df
```
